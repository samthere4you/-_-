```python
import os
import shutil

# Function to organize files in a directory
def organize_files(directory):
    # Define file types and corresponding directories
    file_types = {
        'Images': ['.jpg', '.jpeg', '.png', '.gif'],
        'Documents': ['.pdf', '.doc', '.docx', '.txt'],
        'Videos': ['.mp4', '.avi', '.mov', '.mkv'],
        'Others': []  # Default directory for other file types
    }

    # Create subdirectories if they don't exist
    for folder in file_types.keys():
        folder_path = os.path.join(directory, folder)
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

    # Iterate through files in the directory
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            file_extension = os.path.splitext(filename)[1].lower()

            # Find the appropriate directory for the file type
            target_folder = 'Others'  # Default to 'Others' if type not found
            for folder, extensions in file_types.items():
                if file_extension in extensions:
                    target_folder = folder
                    break

            # Move the file to the corresponding directory
            source_path = os.path.join(directory, filename)
            destination_path = os.path.join(directory, target_folder, filename)
            shutil.move(source_path, destination_path)
            print(f"Moved '{filename}' to '{target_folder}' directory.")

# Main function
if __name__ == "__main__":
    # Specify the directory to organize
    directory_to_organize = 'D:/New folder (2)'

    # Organize files in the specified directory
    organize_files(directory_to_organize)
```
